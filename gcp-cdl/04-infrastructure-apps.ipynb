## Section 3: Containers vs VMs

Containers are lightweight alternatives to VMs for deploying applications.

```python
# Containers vs VMs comparison
comparison = [
    ("Isolation", "Full OS per VM (strong)", "Shared kernel (process-level)"),
    ("Size", "Gigabytes (full OS image)", "Megabytes (app + deps only)"),
    ("Startup Time", "Minutes", "Seconds"),
    ("Portability", "Limited (OS-dependent)", "High (runs anywhere Docker runs)"),
    ("Density", "10s per host", "100s per host"),
    ("Best For", "Legacy apps, full OS control", "Microservices, CI/CD, cloud-native"),
    ("GCP Service", "Compute Engine", "GKE, Cloud Run"),
]

print("Containers vs Virtual Machines")
print("=" * 75)
print(f"{'Aspect':<18} {'VMs':<28} {'Containers'}")
print("-" * 75)
for aspect, vm, container in comparison:
    print(f"{aspect:<18} {vm:<28} {container}")

print("\nDocker Basics:")
print("  Dockerfile  — defines how to build the container image")
print("  Image       — read-only template with app + dependencies")
print("  Container   — running instance of an image")
print("  Registry    — stores images (Artifact Registry on GCP)")
```

## Section 6: Compute Decision Flow

Choose the right compute service based on workload requirements.

```python
# Compute decision flowchart
decisions = [
    {
        "question": "Need full OS control (custom kernel, drivers, GPU)?",
        "yes": "Compute Engine",
        "no": "Continue...",
    },
    {
        "question": "Running many containerized microservices at scale?",
        "yes": "GKE (Kubernetes)",
        "no": "Continue...",
    },
    {
        "question": "Stateless HTTP service or API (container-based)?",
        "yes": "Cloud Run",
        "no": "Continue...",
    },
    {
        "question": "Simple event-triggered function (no long-running)?",
        "yes": "Cloud Functions",
        "no": "Continue...",
    },
    {
        "question": "Traditional web app with built-in scaling?",
        "yes": "App Engine",
        "no": "Re-evaluate requirements",
    },
]

print("Compute Service Decision Flow")
print("=" * 60)
for i, d in enumerate(decisions, 1):
    print(f"\nStep {i}: {d['question']}")
    print(f"  YES -> {d['yes']}")
    print(f"  NO  -> {d['no']}")

print("\nHybrid/Multicloud:")
print("  Need same experience across on-prem + cloud? -> Anthos")
print("  Need high-bandwidth on-prem to GCP connection? -> Cloud Interconnect")
print("  Need encrypted tunnel over internet? -> Cloud VPN")
```

## Section 7: gcloud CLI Examples

Common gcloud commands for infrastructure management.

```python
# Common gcloud commands for infrastructure
commands = {
    "Compute Engine": [
        "gcloud compute instances create my-vm --zone=us-central1-a --machine-type=e2-medium",
        "gcloud compute instances list",
        "gcloud compute instances stop my-vm --zone=us-central1-a",
        "gcloud compute instances delete my-vm --zone=us-central1-a",
    ],
    "GKE": [
        "gcloud container clusters create-auto my-cluster --region=us-central1",
        "gcloud container clusters get-credentials my-cluster --region=us-central1",
        "kubectl get pods",
        "kubectl get services",
    ],
    "Cloud Run": [
        "gcloud run deploy my-service --image=gcr.io/my-project/my-app --region=us-central1",
        "gcloud run services list",
        "gcloud run services describe my-service --region=us-central1",
    ],
    "Cloud Functions": [
        "gcloud functions deploy hello --runtime=python311 --trigger-http --allow-unauthenticated",
        "gcloud functions list",
        "gcloud functions call hello --data='{\"name\": \"CDL\"}'",
    ],
}

print("Common gcloud Commands for Infrastructure")
print("=" * 70)
for service, cmds in commands.items():
    print(f"\n--- {service} ---")
    for cmd in cmds:
        print(f"  $ {cmd}")
```
